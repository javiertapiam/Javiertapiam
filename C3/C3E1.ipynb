```python
df["culmen_length_mm"].fillna(df["culmen_length_mm"].mean(),inplace =True)
df["culmen_depth_mm"].fillna(df["culmen_depth_mm"].mean(),inplace =True)
df["flipper_length_mm"].fillna(df["flipper_length_mm"].mean(),inplace =True)
df["body_mass_g"].fillna(df["body_mass_g"].mean(),inplace =True)
```

```python
def conteo_nulo(x):
  return sum(x.isnull())
df.apply(conteo_nulo,axis=0)
```

```python
from sklearn.model_selection import train_test_split
%matplotlib inline
```

```python
from sklearn.preprocessing import LabelEncoder
df1 = df.copy()
df_sex = ['sex',"species","island"]
label_encoder = LabelEncoder()
for i in df_sex:
    df1[i] = label_encoder.fit_transform(df1[i])
df1.dtypes
```

```python
training_set, test_set = train_test_split(df1.copy(), test_size = 0.3)

print(f'TamaÃ±o set entrenamiento: {len(training_set)}')
print(f'TamaÃ±o set test: {len(test_set)}')
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
features = ["species","island","culmen_length_mm","culmen_depth_mm","flipper_length_mm","body_mass_g","sex"]

training_set[features] = scaler.fit_transform(training_set[features])
test_set[features] = scaler.transform(test_set[features])

display(training_set)
```
