```python
def conteo_nulo(x):
  return sum(x.isnull())
```

```python
df_stats.apply(conteo_nulo,axis=0)
```

Con este conteo sabemos que solo hay 3 columnas con alguna fila sin datos.

Como nos la columna "3PAttempts" nos muestra 517 bloques sin informacion y la cantidad total de bloques es solo un 14% mas de este valor, eliminare la columna debido a que no es un dato tan relevante.

```python
df_stats2 = df_stats.copy()
del df_stats2["3PAttempts"]
```

```python
display(df_stats2)
```

Otra columna con datos faltantes es la columna de "GamesStarted", para llenar esta columna voy a considerar que los jugadores que tengan mas de 35 partidos hayan jugado por lo menos un partido de titular, esto debido a que los valores existentes en la columna de "GamesStarted" solo van del 1 al 0 en el archivo "stats_csv", por lo que los jugadores con menos de 35 partidos tendran 0 juegos de titular.

```python
df_stats2.loc[(df_stats2['Games'] > 35) & (df_stats2['GamesStarted'].isnull()), 'GamesStarted'] = 1
df_stats2['GamesStarted'].fillna(0, inplace=True)
```

Este codigo fue construido con ayuda de inteligencia artificial, en la parte de construccion del codigo, esto debido a que le pedi la notacion de como llenar valores nulos teniendo una condicion. Luego para llenar los valores restantes que no se llenaron debido a la condicion, se llenaron automaticamente con el valor 0 con el comando fillna.

```python
display(df_stats2)
```

Para consolidar estos data frames es necesario juntar la informacion de manera correcta, en donde el salario de cada jugador tiene que corresponder al equipo en el que jugaba en ese entonces, por lo que a la funcion merge le damos las columnas "Player" y "Team" donde solo se consolidad cuando el equipo coincide en los distintos data frames.

```python
df_consolidado = pd.merge(df_stats,df_salaries,on=["Player","Team"])
```
